Model Evaluation using Statsmodel

```python
import statsmodels.formula.api as smf
model1 = smf.ols('Salary~YearsExperience',data=df).fit()
model1.summary()
```

```python
# independent and Dependent variables

x = df[["YearsExperience"]]
y = df["Salary"]
```

# Model Evaluation Using Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
model2.fit(x,y)
```

```python
df["ypred"] = model2.predict(x)
df["ypred"]
```

```python
model2.score(x,y)
```

# Regression Line

```python
sns.scatterplot(x='YearsExperience', y='Salary', data=df, label="Actual Values")
sns.scatterplot(x='YearsExperience', y='ypred', data=df, label="Predicted Values")

# Plot the regression line


plt.plot(df["YearsExperience"],df["ypred"], color="yellow", label='Regression Line')

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

# Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(df["Salary"], df["ypred"])
print(" Mean Squared Error ", mse)
```

# Root Mean Squared Error

```python
rmse = np.sqrt(mean_squared_error(df['Salary'], df['ypred']))
print(" Root Mean Squared Error ",rmse)

```
